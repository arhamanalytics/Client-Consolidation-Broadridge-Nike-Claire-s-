```python
#Adding accesorials
df_accesorial_list=['Additional Handling', 'Address Correction','Audit Fees', 'Billing Adjustment', 'Carbon Neutral', 'Chargebacks','Clearance Fees', 'Collect On Delivery', 'Combined Accessorials','Dangerous Goods', 'Declared Value', 'Delivery Area Fees','Delivery Confirmation', 'Delivery Fees', 'Discounts', 'Duties & Taxes','Invalid Account','Label & Info Fees', 'Large/Oversized Package', 'Late Payment','Manual Shipment', 'Miscellaneous', 'Monitoring Services', 'Non Machinable', 'OverMax/Unauthorized Package', 'Peak Surcharges','Pickup Fees', 'Res/Comm Adjustment', 'Residential Surcharge ','Returns', 'Service Charges', 'Shipping Supplies', 'Tax Forwarding','Third Party ', 'Transportation']
df['Accesorial']=df[df_accesorial_list].sum(axis=1)

#adding totals
df_breakdown=['Freight Fees', 'Fuel Surcharge','Accesorial']
df['Total']=df[df_breakdown].sum(axis=1)

#checking calculated totals with actuals
print(df['Total'].sum(axis=0)-df['Charge Total'].sum(axis=0))
```
